# Regresion simple

**Funcion de coste**

$$RSS = \sum_{i=1}^{n}\left ( y_{i}-\widehat{y^{i}} \right )^{2}$$

### Prediccion
$$\widehat{y} = X\widehat{\beta }$$

### 
El vector de estimadores que minimiza a **RSS**: Error Cuadratico Medio
$$\widehat{\beta } = \left ( X^{T}X \right )^{-1} X^{T}Y$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

Prediccioón del valor de la vivienda respecto con habitaciones promedio

```python
from sklearn.datasets import load_boston # carga la dataset
boston = load_boston()

# Ambas variables son convertidas en matrices con array
X = np.array(boston.data[:,5])# valor promedio de habitaciones por vivienda: variable 5: RM
Y = np.array(boston.target) # valor promedio de la vivienda: variable objetivo

# Graficamos los datos
plt.scatter(X,Y,alpha=0.3) # alpha es transparencia

# se añade una columna de 1 para termino de independientes
X = np.array([np.ones(506),X]).T

# Formula RSS
# Matriz de parametros que minimizan el error cuadratico medio de la nube de puntos
B = np.linalg.inv(X.T @ X) @ X.T @ Y
B # debe mostrar 2 elementos: valor independiente y la pendiente
# Por tanto,-34.67 es el punto del eje y. cuando el x = 0
# la pendiente es de 9.10: cada aumento en una unidad del numero medio de habitaciones, se produce un incremento
# de 9.10 unidades en el valor medio de la vivienda.


# Podemos hacer que la linea roja empiece en 4 y termine en 9.
plt.plot([4,9],[B[0]+B[1]*4, B[0]+B[1]*9],c="red")
plt.show()
```

# Conclusion

Para problemas con muchas variables este metodo resulta poco productivo, ya que el coste de calculo es demasiado.
Ejemplo, si tuvieramos un problema con alta dimensionalidad ( 10.000 atributos), tendriamos que invertir una matriz de 10000x10000, ocacionando un problema de hardware, es decir, las operaciones que tendria que hacer el computador sería de 10000 al cubo operaciones, haciendo que esta forma de encarar el problema sea ineficiente.
Afortunadamente, existe otro metodo para llamado **Gradient Descent**: metodo iterativo que poco a poco va buscando el punto minimo.
